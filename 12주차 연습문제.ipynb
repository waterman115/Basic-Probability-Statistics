```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.array([3, 2, 0, 1, 2, 4, 1, 2, 1, 3, 2, 0, 3, 1, 0]) # 운동시간
y = np.array([15, 21, 25, 23, 19, 20, 23, 19, 20, 17, 21, 28, 20, 22, 26]) # 체질량 지수


cov_xy = np.average((x-np.average(x))*(y-np.average(y)))
print("cov_xy", cov_xy)

plt.scatter(x, y, marker="o", label = "xy", s = 40) # s는 마커의 크기
plt.legend()
plt.xlabel( "x", size = 14)
plt.ylabel( "y", size = 14)

plt.grid()
plt.show()
```

```python
import numpy as np

# 데이터
x = [3, 2, 0, 1, 2, 4, 1, 2, 1, 3, 2, 0, 3, 1, 0]
y = [15, 21, 25, 23, 19, 20, 23, 19, 20, 17, 21, 28, 20, 22, 23]

# 상관계수 계산
correlation = np.corrcoef(x, y)[0, 1]

# 결과 출력
print("Correlation Coefficient:", correlation)
```
